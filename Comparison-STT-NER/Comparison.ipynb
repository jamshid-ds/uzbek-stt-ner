#STT

```python
pipe_base = pipeline("automatic-speech-recognition", model="openai/whisper-base")
pipe_finetuned = pipeline("automatic-speech-recognition", model="jamshidahmadov/whisper-uz-v2")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
list_test = ['test1.wav','test2.wav','test3.wav','test4.wav','test5.wav']

print("Base model natijalari: \n")
for index in range(len(list_test)):
  print(f"{index+1}. {pipe_base(list_test[index])['text']}")

print(' ')

print("Pre trained model natijalari: \n")
for index in range(len(list_test)):
  print(f"{index+1}. {pipe_finetuned(list_test[index])['text']}")
```

#NER

```python
pipe_base = pipeline('ner', model='FacebookAI/xlm-roberta-base', tokenizer='FacebookAI/xlm-roberta-base')
pipe_finetuned = pipeline('ner', model='jamshidahmadov/roberta-ner-uz', tokenizer='jamshidahmadov/roberta-ner-uz')
```

```python
list_test = ["Toshkent shahrida yangi o'zgarishlar bo'lmoqda."]


print("Base model natijalari:\n")
for index, test_sentence in enumerate(list_test, start=1):
    result = pipe_base(test_sentence)  # Asosiy model natijasi
    print(f"Test Sentence: {test_sentence}")
    for item in result:
        print(f"{item['word']} - {item['entity']}")
    print()
print(" ")
print("Pre-trained model natijalari:\n")
for index, test_sentence in enumerate(list_test, start=1):
    result = pipe_finetuned(test_sentence)  # Fine-tuned model natijasi
    print(f"Test Sentence: {test_sentence}")
    for item in result:
        print(f"{item['word']} - {item['entity']}")
    print()
```
